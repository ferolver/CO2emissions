#**Si pudieras revertir el cambio climático ¿cómo lo harías?: Emisiones de CO2**

Las emisiones de dióxido de carbono (CO2) son una de las principales causas del cambio climático, representa el 74% del total de las emisiones de gases de efecto invernadero (Ge, Friedrich y Vigna, 2021). El CO2 se emite a la atmósfera como resultado de la actividad humana, como la quema de combustibles fósiles para producir energía, la deforestación y la agricultura intensiva.

De acuerdo con el artículo publicado por Ge, Friedrich y Vigna (2021), “las emisiones de efecto invernadero (GEI) han aumentado un 50% entre 1990 y 2018”. El aumento de las emisiones de CO2 está causando un calentamiento en todo el planeta, lo que provoca cambios en el clima global, como la elevación del nivel del mar, la acidificación de los océanos, el aumento de la frecuencia y la intensidad de los fenómenos meteorológicos extremos como las sequías, las inundaciones y los huracanes.

Los principales sectores que contribuyen a las emisiones de CO2 son la energía (combustión de combustibles fósiles para la producción de electricidad, calefacción y transporte), la industria, la agricultura y la deforestación. Para reducir las emisiones de CO2, se requieren cambios significativos en la forma en que producimos y consumimos energía, en la forma en que fabricamos y consumimos productos y en la forma en que utilizamos la tierra.

Para abordar el problema de las emisiones de CO2, se han llevado a cabo acuerdos internacionales como el Acuerdo de París, en el que los países se comprometen a reducir sus emisiones de gases de efecto invernadero para limitar el aumento de la temperatura global a menos de 2 grados Celsius por encima de los niveles preindustriales. De acuerdo con la Agenda 2030, "los niveles actuales de emisiones de gases de efecto invernadero deben reducirse a la mitad para 2030 y llegar a cero emisiones netas a mediados de siglo" (Ge, Friedrich y Vigna, 2021).

**El objetivo de este proyecto, es entender cómo está distribuida la aportación de CO2 en el mundo, de tal manera que puedan focalizarse las acciones necesarias para reducir los niveles de CO2 de manera eficiente y estratégica, y plantear soluciones para detener el cambio climático.**

###3. Generación de Emisiones de CO2 entre 1990 y 2019 (período crítico de crecimiento)

```python
lstEmissionsComparative = [ ['1960 to 1989', dfEmissionsByYear[dfEmissionsByYear.index<1990]['MMktonsByYear'].sum()], 
                          ['1990 to 2019', dfEmissionsByYear[dfEmissionsByYear.index>=1990]['MMktonsByYear'].sum()]]

dfEmissionsComparative = pd.DataFrame (lstEmissionsComparative, columns = ['period', 'MMktons'])
```

En la siguiente gráfica se muestra un comparativo de las emisiones generadas en el período 1960-1989 contra las generadas en el período 1990-2019.

```python
f, ax = plt.subplots(figsize=(12, 6)) # Define el tamaño de la figura
plt.title('Comparativo de Emisiones de CO2 1960-1989 vs 1990-2019', fontsize=20)
plt.xlabel('Período', fontsize=12)
plt.ylabel('Emisiones', fontsize=12)
barplt = sns.barplot(data=dfEmissionsComparative, 
                     x=dfEmissionsComparative['period'], 
                     y=dfEmissionsComparative['MMktons'], color='darkblue', 
                     ax=ax).set(xlabel='País', ylabel='Emisiones de CO2 [Millones de kTons]')
```

```python
# Calculamos el % de incremento con respecto al período: 1960 to 1989
fAggegatedInc = ( dfEmissionsComparative[dfEmissionsComparative['period'] =='1990 to 2019']['MMktons'].item() - 
           dfEmissionsComparative[dfEmissionsComparative['period'] =='1960 to 1989']['MMktons'].item() ) * 100 / dfEmissionsComparative[dfEmissionsComparative['period'] =='1960 to 1989']['MMktons'].item()
fAggegatedInc
```

Debido a que China es el principal generador de emisiones de CO2, considerando la muestra de incrementos del período 1990 a 2019, evaluaremos la probabilidad de que reduzca o no incremente más su nivel de generación de CO2 (incremento menor o igual a cero). Esta evaluación la realizaremos por medio de una prueba-T (α = 0.05).

Para ello, planteamos nuestras hipótesis nula y alternativa.


H0: Que el incremento de emisiones de CO2 de China sea a lo más de 0 kTons.

H1: Que el incremento de emisiones de CO2 de China sea mayor a 0 kTons.

```python
#from scipy import stats 
from scipy.stats import ttest_1samp
```

```python
#Ejecutamos la prueba T con la muestra de Incrementos de Emisiones de CO2 de China [1990-2019]
t_stat, p_value = ttest_1samp(dfEmissions1990to2019Ctry['DeltaMMKTons'], popmean=0, alternative='greater')
print("T-statistic value: ", t_stat)  
print("P-Value: ", p_value)
```

Debido a que P <= α, rechazamos la hipótesis nula. Considerando el comportamiento de incrementos en emisiones de CO2 en China entre 1990 a 2019, la probabilidad de que no continúe incrementando su generación de emisiones de CO2 es del **0.0003%**

Esta evaluación la realizaremos por medio de una prueba-T (α = 0.05).

Para ello, planteamos nuestras hipótesis nula y alternativa.


H0: Que el incremento de emisiones de CO2 de USA sea a lo más de 0 kTons.

H1: Que el incremento de emisiones de CO2 de USA sea mayor a 0 kTons.

```python
#from scipy import stats 
from scipy.stats import ttest_1samp
```

```python
#Ejecutamos la prueba T con la muestra de Incrementos de Emisiones de CO2 de China [1990-2019]
t_stat, p_value = ttest_1samp(dfEmissions1990to2019Ctry['DeltaMMKTons'], popmean=0, alternative='greater')
print("T-statistic value: ", t_stat)  
print("P-Value: ", p_value)
```

Debido a que P > α, NO hay elementos para rechazar la hipótesis nula, por lo que, considerando el comportamiento de incrementos en emisiones de CO2 en USA entre 1990 a 2019, no hay elementos para indicar que el nivel de emisiones en USA no se reducirá.
